## Controls

- **Click** a cell to select it
- **Type 1–9** to fill a number
- **Arrow keys** to navigate
- **0, Delete, or Backspace** to clear a cell
- **Q** to close the window

Given (pre-filled) cells are shown in **gray** and cannot be changed.  
Your entries appear in blue. Conflicts are highlighted in red.
